```python
# Dependencies
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt

import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
```

```python
#  Reasd data file that contains incident information by county, resort
incidents_df = pd.read_csv("OutputFiles/Liz/incidentsByCountyAll.csv")
allInfo_df = pd.DataFrame(incidents_df)
allInfo_df = allInfo_df.sort_values(["State","County", "Year"])
```

```python
#  Reformat
allInfo_df = allInfo_df.drop("Unnamed: 0", axis=1)
allInfo_df = allInfo_df.dropna()
allInfo_df = allInfo_df.reset_index()
allInfo_df = allInfo_df.drop("index", axis=1)
```

```python
# Add fatailty rate column and preview to verfiy
allInfo_df["Fatality Rate"] = allInfo_df["Killed"]/allInfo_df["Victims"]*100
allInfo_df.head()
```

```python
# Find all information for different stats using dataframe
#  I was having a problem grouping where a lot of item would show as "nan", so I had to do it this way to not lose data
st = allInfo_df["State"].unique()
loc = {}
yearcolors = {2014:'dimgray' , 2015:'#ff3300' , 2016:'#ff7a33' , 2017:'#ffc266'}

YCOLORS = []
STATE = []
COUNTIES= []
INCIDENTS = []
DENSITIES = []
YEARS = []
FRATES = []
VICTIMS = []

for s in st:
    
    cts = allInfo_df["County"][allInfo_df["State"] == s].unique()
    loc[s] = cts
    std = allInfo_df[allInfo_df["State"] == s]

    for c in range(0, len(cts)):
        
        INC = []
        LABEL = []
        COUNT = []
        DEN = []
        FR = []
        YR = []
        VIC = []
        STT = []
        DN = []

        cd = std[std["County"] == cts[c]]

        for x in range(0, len(cd)):
            i = cd.index[x]
            y = cd["Year"][i]

            LABEL.append(yearcolors[y])
            INC.append(cd[f"Poverty Rate {y}"][i])
            COUNT.append(cd["County"][i])
            VIC.append(cd["Victims"][i])
            FR.append(cd["Fatality Rate"][i])
            DEN.append(cd[f"Density {y}"][i])

        COUNTIES.append(COUNT)
        INCIDENTS.append(INC)
        VICTIMS.append(VIC)
        FRATES.append(FR)    
        YCOLORS.append(LABEL)
        DENSITIES.append(list(DEN))
```

```python
# Plot poverty rates vd victim scatter plot using sets from cell before
with plt.rc_context({'axes.edgecolor':'#262626', 'xtick.color':'#262626', 'ytick.color':'#262626'}):
    fig, ax = plt.subplots()

fig.set_size_inches(10,5, forward=True)

for y in range(0, len(COUNTIES)):
    ax.scatter(INCIDENTS[y], VICTIMS[y], color = YCOLORS[y], s=15)
    
ax.axvline(13, 0, 110, color = 'lightgray', linestyle=":")
lgnd = plt.legend(fontsize = 'small', loc = 'upper right', labels = ["Average US Poverty Rate",2014,2015,2016,2017])

lgnd.legendHandles[1].set_color('dimgray')
lgnd.legendHandles[2].set_color('#ff3300')
lgnd.legendHandles[3].set_color('#ff7a33')
lgnd.legendHandles[4].set_color('#ffc266')

# ax.tick_params(axis='y', left=False, labelleft=False)
ax.set_ylim(-0.25, 100)
ax.set_ylabel("Victims Per Incident", color = '#262626', labelpad = 5, fontsize = 12)
ax.set_xlabel("Poverty Rate (%)", color = '#262626', labelpad = 5, fontsize = 12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.title(f"Poverty Rate Vs Victims (County)", fontsize=18, color = '#262626')
plt.tight_layout()

# plt.savefig("Images/Liz/PovertyRateVsTotalVictimsPerCounty")
plt.show()
```

```python
# Plot Poverty Rate vs Fatality Rate for totals by counties
with plt.rc_context({'axes.edgecolor':'#262626', 'xtick.color':'#262626', 'ytick.color':'#262626'}):
    fig, ax = plt.subplots()

fig.set_size_inches(10,5, forward=True)

for y in range(0, len(COUNTIES)):
    ax.scatter(INCIDENTS[y], FRATES[y], color = YCOLORS[y], s=15)
    
ax.axvline(13, 0, 110, color = 'lightgray', linestyle=":", label="Avg")
lgnd = plt.legend(fontsize = 'small', loc = 'upper right', labels = ["Average US Poverty Rate",2014,2015,2016,2017])

lgnd.legendHandles[1].set_color('dimgray')
lgnd.legendHandles[2].set_color('#ff3300')
lgnd.legendHandles[3].set_color('#ff7a33')
lgnd.legendHandles[4].set_color('#ffc266')

ax.set_ylim(-0.25, 100)
ax.set_ylabel("Fatality Rate (%)", color = '#262626', labelpad = 5, fontsize = 12)
ax.set_xlabel("Poverty Rate (%)", color = '#262626', labelpad = 5, fontsize = 12)
ax.set_ylim(-5, 105)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.title(f"Poverty Rate Vs Fatality Rate Per County", fontsize=18, color = '#262626')
plt.tight_layout()

plt.savefig("Images/Liz/PovertyRateVsFatalityRatePerCounty")
plt.show()
```

### Quick regression model test

```python
# Narrow down dataframe to use in quick regression model
allInfo_df = allInfo_df[["State","County","Incidents","Injured","Killed","Victims","Population 2017", "Density 2017","Poverty Rate 2017", "Region", "Fatality Rate"]]
allInfo_df = allInfo_df.rename(columns={"Density 2017":"Density", "Population 2017":"Population", "Poverty Rate 2017":"Poverty Rate"})

# Define density type to use as dummy variable
for x in range(0, len(allInfo_df)):
    
    if allInfo_df["Density"][x] > 3000:
        allInfo_df.loc[x, "Type"] = "Urban"
        
    elif 1000 < allInfo_df["Density"][x] < 3000:
        allInfo_df.loc[x, "Type"] = "Suburban"
        
    else:
        allInfo_df.loc[x, "Type"] = "Rural"
        
allInfo_df.head()
```

```python
# Calculate Per Capita Amounts
allInfo_df["Injured/Capita"] = allInfo_df["Injured"]/allInfo_df["Population"]*100000
allInfo_df["Incidents/Capita"] = allInfo_df["Incidents"]/allInfo_df["Population"]*100000
allInfo_df["Fatalities/Capita"] = allInfo_df["Killed"]/allInfo_df["Population"]*100000
allInfo_df["Victims/Capita"] = allInfo_df["Victims"]/allInfo_df["Population"]*100000
allInfo_df.head()
```

```python
# Sort through data to find information for multiple incidents in each unique county, state
# fatalityRates = []
# victimSize = []

st = list(allInfo_df["State"].unique())
loc = {}
xmin = []
for s in st:
    cts = allInfo_df["County"][allInfo_df["State"] == s].unique()
    loc[s] = cts

COUNTIES_M= []
INCIDENTS_M = []
VICTIMS_M = []
FRATES_M = []
TYPES_M = []
STATEM = []
VICTIMS_M = []
TCOLORS_M = []


tcolors = { "Rural":'dimgray' , "Suburban":'#ff7a33' , "Urban":'#ffc266'}

for s in st:
    
    cts = loc[s]
    std = allInfo_df[allInfo_df["State"] == s]

    for c in cts:
        
        cd = std[std["County"] == c]        

        for x in range(0, len(cd)):
            i = cd.index[x]
            t = cd["Type"][i]
            COUNTY = []
            INC = []
            VIC = []        
            COLOR = []
            FRT = []
            TYPE = []
            STATE = []

            if cd["Incidents"][i]>1:

                COLOR.append(tcolors[t])
                TCOLORS_M.append(COLOR)

                VIC.append(cd["Victims"][i])
                VICTIMS_M.append(VIC)
        
                INC.append(cd["Incidents"][i])
                INCIDENTS_M.append(INC)
    
                COUNTY.append(cd["County"][i])
                COUNTIES_M.append(COUNTY)
            
                FRT.append(round(cd["Fatality Rate"][i], 2))
                FRATES_M.append(FRT)
                
                TYPE.append(cd["Type"][i])
                TYPES_M.append(TYPE)
                
                STATE.append(cd["State"][i])
                STATEM.append(STATE)
```
