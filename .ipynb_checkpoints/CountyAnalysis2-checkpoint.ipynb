```python
allInfo_df = allInfo_df[["State","County","Incidents","Injured","Killed","Victims","Population 2017", "Density 2017","Poverty Rate 2017", "Region", "Fatality Rate"]]
allInfo_df = allInfo_df.rename(columns={"Density 2017":"Density", "Population 2017":"Population", "Poverty Rate 2017":"Poverty Rate"})

# Define density type to use as dummy variable
for x in range(0, len(allInfo_df)):
    
    if allInfo_df["Density"][x] > 3000:
        allInfo_df.loc[x, "Type"] = "Urban"
        
    elif 1000 < allInfo_df["Density"][x] < 3000:
        allInfo_df.loc[x, "Type"] = "Suburban"
        
    else:
        allInfo_df.loc[x, "Type"] = "Rural"
        
allInfo_df.head()
```

```python
allInfo_df["Injured/Capita"] = allInfo_df["Injured"]/allInfo_df["Population"]*100000
allInfo_df["Incidents/Capita"] = allInfo_df["Incidents"]/allInfo_df["Population"]*100000
allInfo_df["Fatalities/Capita"] = allInfo_df["Killed"]/allInfo_df["Population"]*100000
allInfo_df["Victims/Capita"] = allInfo_df["Victims"]/allInfo_df["Population"]*100000
```

```python
allInfo_df.head()
```

```python
countyByType = allInfo_df.groupby("Type")
countyByType_df = countyByType.agg({
    "Incidents":['sum'], "Injured":['sum'], "Killed":['sum'], "Victims":['sum'], "Density":['mean'], 
    "Poverty Rate":['mean'], "Fatality Rate":['mean'], "Incidents/Capita":['mean'], "Fatalities/Capita":['mean']
})
countyByType_df = countyByType_df.rename(columns={"Incidents":"Total Incidents","Injured":"Total Injured",
                                                  "Killed":"Total Fatalities","Victims":"Total Victims",
                                                  "Density":"Average Density","Poverty Rate":"Average Poverty Rate",
                                                  "Fatality Rate":"Average Fatality Rate", 
                                                  "Incidents/Capita":"Avergae Incidents Per Capita", 
                                                  "Fatalities/Capita":"Average Fatalities Per Capita"
                                                 })
countyByType_df.columns = countyByType_df.columns.droplevel(1)
countyByType_df.head()
```

```python
fatalRate = countyByType_df["Total Fatalities"]/countyByType_df["Total Victims"]*100
fatalRate = round(fatalRate, 2)
```
