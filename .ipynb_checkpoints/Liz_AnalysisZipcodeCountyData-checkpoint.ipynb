### Calculate per capita by zipcode

```python
# Calculate per capita numbers
incidents_df["Victims Per Capita"] = incidents_df["Victims"] / incidents_df["Population At Incident"] * 1000
incidents_df["Victims Per Capita"] = round(incidents_df["Victims Per Capita"], 2)

incidents_df["Fatalities Per Capita"] = incidents_df["Killed"] / incidents_df["Population At Incident"] * 1000
incidents_df["Fatalities Per Capita"] = round(incidents_df["Fatalities Per Capita"], 2)

# Preview to verify
incidents_df.head()
```

```python
# Calulate days of year for use as x-axis in graph
daysOfMonth = {}
daysOfMonthL = {}
months = [f for f in np.arange(1, 13, 1)]

length = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
lengthL = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

monthTotal = 0
monthTotalL = 0
monLength = []
monLengthL = []

for x in range(0, len(length)):
    monthTotal = monthTotal + length[x]
    monLength.append(monthTotal)
    monthTotalL = monthTotalL + lengthL[x]
    monLengthL.append(monthTotalL)
    
for x in range(0, len(months)):
    daysOfMonth[x+1] = monLength[x]
    daysOfMonthL[x+1] = monLengthL[x]

DayDate = []
for y in range(0, len(incidentDates_df)):
    yr = int(incidentDates_df["Year"][y])
    mo = int(incidentDates_df["Month"][y])
    dy = int(incidentDates_df["Day"][y])

    if yr == 2016:
        DayDate.append(daysOfMonthL[mo] + dy)
    else:
        DayDate.append(daysOfMonth[mo] + dy)

incidentDates_df["TimeInDays"] = DayDate
incidentDates_df.head()
```

```python
# Define density type to use as dummy variable
for x in range(0, len(dataSet)):
    
    if dataSet["Density"][x] > 3000:
        dataSet.loc[x, "Type"] = "Urban"
        
    elif 1000 < dataSet["Density"][x] < 3000:
        dataSet.loc[x, "Type"] = "Suburban"
        
    else:
        dataSet.loc[x, "Type"] = "Rural"
```

```python
# Group dataframe by state and county and calculate 
grouped = dataSet.groupby(['State', 'County'])
grouped = grouped.agg({
    "Incidents":['sum'], "Injured":['sum'],"Killed":['sum'], "Victims":['sum'], 
    "Den Change Avg (%)":['last'], "Poverty Rate Change Avg (%)":['last'], "Population":['mean'], 
    "Type":['last'], "Region":['last'], "Pop Change Avg (%)":['last'], "Poverty Rate":['mean'], "Density":['mean']})
grouped.columns = grouped.columns.droplevel(1)

# Preview
grouped.head()
```

```python
# Calculate per capita values
grouped["Incidents Per Capita"] = grouped["Incidents"]/grouped["Population"]*100000
grouped["Fatalities Per Capita"] = grouped["Killed"]/grouped["Population"]*100000
grouped["Fatality Rate"] = grouped["Killed"]/grouped["Victims"]*100
```

```python
# Drop na rows
grouped = grouped.dropna()

# Create dummy variable for density type
dummies = pd.get_dummies(grouped['Type'])
grouped = pd.concat([grouped, dummies], axis=1)

# Create dummy variables for region
dummies2 = pd.get_dummies(grouped['Region'])
grouped = pd.concat([grouped, dummies2], axis=1)

# View to verify
grouped.columns
```

```python
# Regression model with Incidents Per Capita as dependent variable
# x_data = stab[['Poverty Rate Change Avg (%)', 'Den Change Avg (%)', 'Rural', 'Suburban', 'Urban']]
x_data = grouped[['Rural', "Poverty Rate"]]

y_data = grouped['Incidents Per Capita']

# Add constant term to the linear regression model
x_data = sm.add_constant(x_data)

# Run regression model
model = sm.OLS(y_data, x_data).fit()
predictions = model.predict(x_data)

# Print results
result = model.summary()
print(result)
```
